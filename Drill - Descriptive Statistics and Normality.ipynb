```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exponential = np.random.exponential(5,100)
rayleigh = np.random.rayleigh(5,100)
power = np.random.power(5,100)
gamma = np.random.standard_gamma(5,100)
weibull = np.random.weibull(5,100)
logistic = np.random.logistic(5,5,100)
```

```python
plt.figure(figsize=(30,30))

plt.subplot(321)
plt.hist(exponential, bins=15, color='c')
plt.title('Exponential',size=55)
# Add vertical lines
plt.axvline(exponential.mean(), color='black', linestyle='solid', linewidth=5)
plt.axvline(exponential.mean() + exponential.std(), color='black', linestyle='dashed', linewidth=3.5)
plt.axvline(exponential.mean() - exponential.std(), color='black', linestyle='dashed', linewidth=3.5) 


plt.subplot(322)
plt.hist(rayleigh, bins=15, color='r')
plt.title('Rayleigh',size=55)
# Add vertical lines
plt.axvline(rayleigh.mean(), color='black', linestyle='solid', linewidth=5)
plt.axvline(rayleigh.mean() + rayleigh.std(), color='black', linestyle='dashed', linewidth=3.5)
plt.axvline(rayleigh.mean() - rayleigh.std(), color='black', linestyle='dashed', linewidth=3.5) 


plt.subplot(323)
plt.hist(power, bins=15, color='b')
plt.title('Power',size=55)
# Add vertical lines
plt.axvline(power.mean(), color='black', linestyle='solid', linewidth=5)
plt.axvline(power.mean() + power.std(), color='black', linestyle='dashed', linewidth=3.5)
plt.axvline(power.mean() - power.std(), color='black', linestyle='dashed', linewidth=3.5) 


plt.subplot(324)
plt.hist(gamma, bins = 15, color='m')
plt.title('Gamma',size=55)
# Add vertical lines
plt.axvline(gamma.mean(), color='black', linestyle='solid', linewidth=5)
plt.axvline(gamma.mean() + gamma.std(), color='black', linestyle='dashed', linewidth=3.5)
plt.axvline(gamma.mean() - gamma.std(), color='black', linestyle='dashed', linewidth=3.5) 


plt.subplot(325)
plt.hist(weibull, bins=15, color='g')
plt.title('Weibull',size=55)
# Add vertical lines
plt.axvline(weibull.mean(), color='black', linestyle='solid', linewidth=5)
plt.axvline(weibull.mean() + weibull.std(), color='black', linestyle='dashed', linewidth=3.5)
plt.axvline(weibull.mean() - weibull.std(), color='black', linestyle='dashed', linewidth=3.5)


plt.subplot(326)
plt.hist(logistic, bins=15, color='y')
plt.title('Logistic',size=55)
# Add vertical lines
plt.axvline(logistic.mean(), color='black', linestyle='solid', linewidth=5)
plt.axvline(logistic.mean() + logistic.std(), color='black', linestyle='dashed', linewidth=3.5)
plt.axvline(logistic.mean() - logistic.std(), color='black', linestyle='dashed', linewidth=3.5)


plt.tight_layout()
plt.show()
```

## Observation

The mean (and therefore the standard deviation) is virtually useless for distributions that are intensely skewed left or right, as the numbers on either end will drag the mean up or down so that it no longer communicates the true center of the data (*see Exponential and Power distributions*)

However, in certain distributions that were skewed the mean was useful IF there was a less extreme skew (*see Gamma and Rayleigh distribution*)

In all of them however, the combination of mean and standard deviation approximated the majority of the data.

```python
rand1 = np.random.normal(5,.5,100)
rand2 = np.random.normal(10,1,100)
rand3 = rand1 + rand2
mean = np.mean(rand3)
std = np.std(rand3)
```

```python
plt.hist(rand3,bins=20)
plt.axvline(mean, color='black', linestyle='solid', linewidth=2)
plt.axvline(mean + std, color='black', linestyle='dashed', linewidth=2)
plt.axvline(mean - std, color='black', linestyle='dashed', linewidth=2) 
```

```python
print(mean)
print(std)
```

## Observation

When you combine 2 normal distributions A and B, the resulting distribution will have mean A.mean() + B.mean() and standard deviation of np.sqrt(A.var() + B.var())
